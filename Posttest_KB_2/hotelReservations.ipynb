# Analisis Data Reservasi Hotel

### Import Library

```python
import pandas as pd
import prettytable as prtl
```

### Import dataset

```python
df = pd.read_csv("Hotel Reservations.csv")
```

### Melihat bentuk dari dataset

```python
df.info()
```

### Menampilkan statistik dari attribute bertipe numerik

```python
df.describe()
```

### Menampilkan 10 record pertama dari dataset dengan attribute numerik

```python
df.select_dtypes(include=['int64', 'float64']).head(10)
```

### Menampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe data numerik

```python
datasetInformation = {}

datasetInformation['Jumlah Record'] = df.shape[0]
datasetInformation['Jumlah Attribute'] = df.shape[1]
datasetInformation['Nama Attribute'] = df.columns.tolist()
datasetInformation['Jumlah Attribute Dengan Tipe Angka'] = df.select_dtypes(include=['int64', 'float64']).shape[1]

dictKeys = list(datasetInformation.keys())

table = prtl.PrettyTable()
table.add_column(dictKeys[0], [datasetInformation['Jumlah Record']])
table.add_column(dictKeys[1], [datasetInformation['Jumlah Attribute']])
table.add_column(dictKeys[2], ['\n'.join(datasetInformation['Nama Attribute'])])
table.add_column(dictKeys[3], [datasetInformation['Jumlah Attribute Dengan Tipe Angka']])
print(table)

```

### Menampilkan nilai dataset pada baris dimulai dengan 3 digit NIM terakhir (032) hingga data terakhir dan kolom 4 hingga kolom terakhir

```python
df.iloc[32:, 3:] # menggunakan iloc
```

```python
# menggunakan loc
df.loc[32:, datasetInformation['Nama Attribute'][3:]]
```
