### Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import Dataset

```python
df = pd.read_csv('Hotel Reservations.csv')
```

### Melihat bentuk dari dataset

```python
df.info()
```

### Melihat 10 Data Teratas

```python
df.head(10)
```

### Melihat korelasi antar kolom pada dataset

```python
plt.figure(figsize=(15, 5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="flare")
plt.show()
```

### Jumlah Tiap Status Booking pada Hotel

```python
dfBookingStatus = df['booking_status'].value_counts()
plt.figure(figsize=(15,5))
plt.barh(dfBookingStatus.index, dfBookingStatus.values, color=['blue', 'red'])
plt.xlabel('Banyak Booking')
plt.ylabel('Status Booking')
plt.title('Perbandingan Status Booking pada Hotel')
plt.show()
```

### Perbandingan Jumlah Booking tiap Metode berdasarkan Status Booking

```python
sns.set_style("darkgrid")
plt.figure(figsize=(15, 5))
sns.countplot(data=df, x="market_segment_type", palette="Set2", hue="booking_status")
plt.xlabel("Jenis Booking")
plt.ylabel("Jumlah Booking")
plt.title("Perbandingan Jumlah Booking tiap Metode berdasarkan Status Booking", fontsize=20, fontweight="bold")
plt.show()
```

### Perbandingan Status Booking berdasarkan Lama Waktu Tunggu

```python
plt.figure(figsize=(15, 5))
sns.histplot(data=df, x="lead_time", hue="booking_status", multiple="stack", palette="Set2", bins=100)
plt.xlabel("Waktu Tunggu")
plt.ylabel("Jumlah")
plt.title("Perbandingan Status Booking berdasarkan Lama Waktu Tunggu", fontsize=20, fontweight="bold")
plt.show()
```

### Banyaknya Booking berdasarkan Bulan

```python
plt.figure(figsize=(15, 5))
sns.set_style("darkgrid")
sns.countplot(data=df, x="arrival_month", palette="Set2")
plt.xlabel("Bulan Kedatangan")
plt.ylabel("Jumlah Booking")
plt.title("Perbandingan Bulan Kedatangan", fontsize=20, fontweight="bold")
plt.show()
```

### Perbandingan Status Booking setiap Bulan

```python
plt.figure(figsize=(15, 5))
sns.set_style("darkgrid")
sns.countplot(data=df, x="arrival_month", palette="Set2", hue="booking_status")
plt.xlabel("Bulan Kedatangan")
plt.ylabel("Jumlah Booking")
plt.title("Perbandingan Bulan Kedatangan berdasarkan Status Booking", fontsize=20, fontweight="bold")
plt.show()
```

### Perbandingan Jumlah tiap Jenis Makanan

```python
plt.figure(figsize=(15, 5))
sns.set_style("darkgrid")
sns.countplot(data=df, x="type_of_meal_plan", palette="Set2")
plt.xlabel("Jenis Makanan")
plt.ylabel("Jumlah Booking")
plt.title("Perbandingan Jenis Makanan", fontsize=20, fontweight="bold")
plt.show()
```

### Persentase Jenis Makanan yang Dipesan

```python
plt.figure(figsize=(15, 5))
sns.set_theme()
plt.pie(df['type_of_meal_plan'].value_counts(), labels=df['type_of_meal_plan'].value_counts().index, autopct='%.2f%%', shadow=True, startangle=90, explode=[0, 0.1, 0, 0])
plt.title("Persentase Jenis Makanan", fontsize=20, fontweight="bold")
plt.show()
```

### Perbandingan Status Booking berdasarkan Jenis Makanan

```python
plt.figure(figsize=(15, 5))
sns.set_style("darkgrid")
sns.countplot(data=df, x="type_of_meal_plan", palette="Set2", hue="booking_status")
plt.xlabel("Jenis Makanan")
plt.ylabel("Jumlah Booking")
plt.title("Perbandingan Status Booking berdasarkan Jenis Makanan", fontsize=20, fontweight="bold")
plt.show()
```

```python
plt.figure(figsize=(15, 5))
sns.set_style("darkgrid")
sns.countplot(data=df, x="room_type_reserved", palette="Set2", hue="booking_status")
plt.xlabel("Tipe Kamar")
plt.ylabel("Jumlah Booking")
plt.title("Perbandingan Jumlah Booking tiap Tipe Kamar berdasarkan Status Booking", fontsize=20, fontweight="bold")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.hist(df["avg_price_per_room"], bins=20, color="green")
plt.xlabel("Harga Rata-rata per Kamar")
plt.ylabel("Jumlah")
plt.title("Distribusi Harga Rata-rata per Kamar", fontsize=20, fontweight="bold")
plt.show()
```

```python
df['date'] = pd.to_datetime(df['arrival_year'].astype(str) + '-' + df['arrival_month'].astype(str) + '-' + df['arrival_date'].astype(str), errors='coerce')

plt.figure(figsize=(15, 5))
sns.set_style("darkgrid")
sns.lineplot(data=df, x="date", y="avg_price_per_room", hue="booking_status", palette="Set2")
plt.xlabel("Tanggal")
plt.ylabel("Rata-rata Harga per Kamar")
plt.title("Trend Rata-rata Harga per Kamar berdasarkan tanggal", fontsize=20, fontweight="bold")
plt.show()
```

## Kesimpulan
Hal-hal yang berdampak terhadap status booking hotel terdiri dari:
1. Lama waktu tunggu, semakin lama waktu tunggu maka semakin besar kemungkinan booking dibatalkan.
2. Harga kamar, semakin mahal harga kamar maka semakin besar kemungkinan booking dibatalkan.

Kesimpulan lainnya:
1. Jika dilihat dari korelasi antar kolom, kolom yang kemungkinan mempengaruhi harga rata-rata kamar adalah banyaknya orang dewasa dan anak-anak yang menginap.
2. Jenis makanan yang banyak dipesan adalah "Meal Plan 1".
3. Jenis kamar yang banyak dipesan adalah "Room_Type 1".
4. Hampir setiap bulannya, booking yang di cancel setengah dari booking yang tidak di cancel.
5. Bulan yang paling banyak booking adalah bulan Oktober.
6. Harga terendah kamar berada pada bulan Agustus 2017, sedangkan yang tertinggi ada di akhir bulan September 2017.
7. Metode booking yang paling banyak digunakan adalah "Online".
